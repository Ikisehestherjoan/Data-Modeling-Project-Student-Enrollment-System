# 🗄️ **SQL: CREATE DATABASE + TABLES (Student Enrollment System)**

## 🔹 1. Create Database

```sql
CREATE DATABASE student_enrollment;

---

# 📘 **2. CREATE TABLES**

---

## 🟦 STUDENT

```sql
CREATE TABLE Student (
    student_id INT PRIMARY KEY,
    name VARCHAR(100),
    email VARCHAR(100),
    phone VARCHAR(20),
    dob DATE
);
```

---

## 🟩 INSTRUCTOR

```sql
CREATE TABLE Instructor (
    instructor_id INT PRIMARY KEY,
    name VARCHAR(100),
    department VARCHAR(100),
    email VARCHAR(100)
);
```

---

## 🟦 COURSE

```sql
CREATE TABLE Course (
    course_id INT PRIMARY KEY,
    course_code VARCHAR(50),
    title VARCHAR(150),
    credits INT,
    semester VARCHAR(50),
    instructor_id INT,
    FOREIGN KEY (instructor_id) REFERENCES Instructor(instructor_id)
);
```

---

## 🟨 ENROLLMENT (Bridge Table)

```sql
CREATE TABLE Enrollment (
    enrollment_id INT PRIMARY KEY,
    student_id INT,
    course_id INT,
    enrollment_date DATE,
    grade VARCHAR(10),
    FOREIGN KEY (student_id) REFERENCES Student(student_id),
    FOREIGN KEY (course_id) REFERENCES Course(course_id)
);
```

---

## 🟪 SPONSOR

```sql
CREATE TABLE Sponsor (
    sponsor_id INT PRIMARY KEY,
    name VARCHAR(100),
    organization VARCHAR(150),
    contact VARCHAR(100)
);
```

---

## 🟥 PAYMENT

```sql
CREATE TABLE Payment (
    payment_id INT PRIMARY KEY,
    student_id INT,
    enrollment_id INT,
    amount DECIMAL(10,2),
    payment_date DATE,
    status VARCHAR(50),
    sponsor_id INT NULL,
    FOREIGN KEY (student_id) REFERENCES Student(student_id),
    FOREIGN KEY (enrollment_id) REFERENCES Enrollment(enrollment_id),
    FOREIGN KEY (sponsor_id) REFERENCES Sponsor(sponsor_id)
);
```

---

# 🔗 **RELATIONSHIPS IMPLEMENTED (From my ERD)**

* Student → Enrollment (1:N) ✅
* Course → Enrollment (1:N) ✅
* Instructor → Course (1:N) ✅
* Enrollment → Payment (1:N) ✅
* Sponsor → Payment (1:N, optional) ✅
* Student → Payment (1:N) ✅
